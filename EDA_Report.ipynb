```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('RT_DA_Task_5_Data.csv')
df.head()
```

```python
summary = df.describe()
print(summary)
```

```python
df = df.dropna()
df.head(5)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], kde=True)
plt.title('Age split')
plt.show()
print("average age is",df['Age'].mean())
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Monthly Income', data=df)
plt.title('Income split')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(y='Age', x='Monthly Income', data=df)
plt.title('plot1')
plt.xlabel('income')
plt.ylabel('avg age count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(y='Family size', x='Monthly Income', data=df)
plt.title('plot2')
plt.xlabel('income')
plt.ylabel('average family member count')
plt.show()
```
